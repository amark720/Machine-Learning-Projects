```python
df = pd.read_csv("StarWars.csv",skiprows=2,sep=',',header=None) #Import the dataset excluding first 2 rows and Headers.
column_name =pd.read_csv("StarWars.csv") #Importing this separately to view the column name
pd.set_option('display.max_rows', df.shape[0]+1) # To display the whole rows without Rows Fold
pd.set_option('display.max_columns', None) #TO display the whole columns without display fold
```

```python
df1.drop([0, 38, 39], axis=1, inplace=True) #Dropping column number 0, 38, 39 as it's of no use
df1.head(5)
```

```python
#Performing data cleaning of Age Column
print("Before Data cleaning:\n",df1["Age"].value_counts(dropna = False))
df1["Age"] = df1["Age"].fillna('45-60')
df1["Age"] = df1["Age"].str.replace('500', '30-44')
print("\nAfter Data cleaning:")
df1["Age"].value_counts(dropna = False)
```

```python
#Plotting a Pie Chart for Age Distribution
Age=pd.DataFrame()
Age['Type']=df1['Age'].value_counts().index
Age['Count']=df1['Age'].value_counts().values
py.init_notebook_mode(connected=True)
fig = go.Figure(data=[go.Pie(labels=Age['Type'], values=Age['Count'],hole=0.2)])
fig.update_layout(title_text='Age Group Involved into this survey!')
fig.show()
```

```python
#Displaying state whether user view the following characters favorably, unfavorably, or are unfamiliar with him/her for Obi Wan Kenobi
df1["Obi Wan Kenobi Fans"].value_counts(dropna = False)
```

```python
# set width of bar
barWidth = 0.10
plt.figure(figsize=(20,10))
 
# set height of bar
bars1 = [610, 552, 547, 245, 591] # bars value filled after calculating total count of each category according to above code cell
bars2 = [151, 219,210 , 269, 159]
bars3 = [44, 38, 48, 135, 43]
bars4 = [8, 13, 12, 83, 8]
bars5 = [1, 3, 6, 39, 7]
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
r5 = [x + barWidth for x in r4]
 
# Make the plot
plt.bar(r1, bars1, color='#3339FF', width=barWidth, edgecolor='white', label='Very favorably')
plt.bar(r2, bars2, color='#557f2d', width=barWidth, edgecolor='white', label='Somewhat favorably')
plt.bar(r3, bars3, color='#2d7f5e', width=barWidth, edgecolor='white', label='neutral')
plt.bar(r4, bars4, color='#FF33DA', width=barWidth, edgecolor='white', label='Somewhat unfavorably')
plt.bar(r5, bars5, color='#7f6d5f', width=barWidth, edgecolor='white', label='Very unfavorably')
 
# Add xticks on the middle of the group bars
plt.xlabel('group', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], ['Han Solo', 'Luke Skywalker', 'Princess Leia Organa', 'Anakin Skywalker', 'Obi Wan Kenobi'])
 
# Create legend & Show graphic
plt.legend()
plt.title("Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.", y=1.02);
plt.show()
```

```python
#Performing data cleaning of Gender column.
print("Before Data cleaning:\n",df1["Gender"].value_counts(dropna = False))
df1["Gender"] = df1["Gender"].str.replace('Female', 'female')
df1["Gender"] = df1["Gender"].str.replace('F', 'female')
print("\nAfter Data cleaning:")
df1["Gender"].value_counts(dropna = True)
```

```python
#Plotting the graph of people according to their gender.
sns.set(font_scale=1.4)
df1["Gender"].value_counts().plot(kind='bar', figsize=(7, 6), rot=10)
plt.xlabel("Gender", labelpad=14)
plt.ylabel("Count of People", labelpad=14)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) 
plt.title("Count of People Who Received Tips by Gender", y=1.02);
```

```python
# Plotting Star Wars Episodes Ratings: 0 means Worst-Liked and 6 means most-Liked
import plotly.graph_objects as go
from plotly.subplots import make_subplots
py.init_notebook_mode(connected=True)
#df1['Rating: Star Wars: Episode I The Phantom Menace'].value_counts().sort_index(ascending=False)
labels = [0,1]
fig = make_subplots(3, 2, specs=[[{'type':'domain'}, {'type':'domain'}],[{'type':'domain'}, {'type':'domain'}],[{'type':'domain'}, {'type':'domain'}]],
                    subplot_titles=['Episode I The Phantom Menace', 'Episode II Attack of the Clones',
                                    'Episode III Revenge of the Sith','Episode IV A New Hope','Episode VI Return of the Jedi'])

fig.add_trace(go.Pie(labels=df1['Rating: Star Wars: Episode I The Phantom Menace'].index, values=df1['Rating: Star Wars: Episode I The Phantom Menace'].value_counts().sort_index(ascending = False), scalegroup='two',
                     name="Success Rate"), 1, 1)

fig.add_trace(go.Pie(labels=df1['Rating: Star Wars: Episode II Attack of the Clones'].index, values=df1['Rating: Star Wars: Episode II Attack of the Clones'].value_counts().sort_index(ascending = False), scalegroup='two',
                     name="Success Rate"), 1, 2)
fig.add_trace(go.Pie(labels=df1['Rating: Star Wars: Episode III Revenge of the Sith'].index, values=df1['Rating: Star Wars: Episode III Revenge of the Sith'].value_counts().sort_index(ascending = False), scalegroup='two',
                     name="Success Rate"), 2, 1)
fig.add_trace(go.Pie(labels=df1['Rating: Star Wars: Episode IV A New Hope'].index, values=df1['Rating: Star Wars: Episode IV A New Hope'].value_counts().sort_index(ascending = False), scalegroup='two',
                     name="Success Rate"), 2, 2)
fig.add_trace(go.Pie(labels=df1['Rating: Star Wars: Episode VI Return of the Jedi'].index, values=df1['Rating: Star Wars: Episode VI Return of the Jedi'].value_counts().sort_index(ascending = False), scalegroup='two',
                     name="Success Rate"), 3, 1)

fig.update_layout(margin=dict(t=100, b=10, l=20, r=10))
fig.update_layout(title_text='Star Wars Episodes Ratings: 0 means Worst Liked and 6 means most Liked',height = 800,
                   width = 800,
                   autosize = False)

fig.show()
```
